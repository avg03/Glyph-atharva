```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Introduction to Matplotlib

### 1. Basic Line Plot: Tracking Data Over Time

```python
# Data: Time steps (X) vs. Measurement (Y)
time_steps = np.arange(1, 6)
measurement_data = np.array([5, 8, 12, 15, 18])

# Create the plot
plt.figure(figsize=(6, 4))
plt.plot(time_steps, measurement_data)

# Add labels, title, and legend
plt.xlabel("Time Steps")
plt.ylabel("Measurement Value")
plt.title("Data Trend Over Time")

# Show the plot
plt.show()
```

##### Multiple lines on one plot

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Plot multiple lines
plt.plot(x, y1, label="y = x^2", color="blue", linestyle="--", marker="o")
plt.plot(x, y2, label="y = x", color="red", linestyle="-.", marker="s")

# Add labels, legend, and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()

# Show the plot
plt.show()
```

### 2. Histogram

```python
# Data: 500 random samples from a normal distribution
sample_data = np.random.normal(loc=10, scale=2, size=500)

# Create histogram
plt.figure(figsize=(7, 5))
plt.hist(sample_data, 
         bins=25, # Number of bars
         color='lightcoral', 
         edgecolor='black')

plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Data (Normal)")
plt.show()
```

### 3. Scatter Plot

```python
# Data
variable_a = np.arange(10)
variable_b = variable_a + np.random.uniform(low=-1.5, high=1.5, size=10)
#notice that 'b' is 'a' added with a random noise

# Create scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(variable_a, variable_b, 
            color='forestgreen', 
            marker='o')

plt.xlabel("Variable A")
plt.ylabel("Variable B")
plt.title("Correlation between A and B")
plt.show()
```

### 4. Bar Chart

```python
# Data
categories = ['Project A', 'Project B', 'Project C', 'Project D']
time_spent = [15, 22, 10, 30]

# Create bar chart
plt.figure(figsize=(7, 5))
plt.bar(categories, time_spent, color=['skyblue', 'salmon', 'lightgreen', 'gold'])

plt.xlabel("Project")
plt.ylabel("Hours Spent")
plt.title("Time Allocation per Project")
plt.show()
```

### 5. Subplots (Multiple visualizations in one plot)

```python
# Data
data_set_1 = np.random.rand(100)
data_set_2 = np.random.randint(0, 10, 10)
categories_2 = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10']

# Create a figure
plt.figure(figsize=(12, 7))

# 1 row and 2 columns
# --- First Subplot (1, 2, 1) ---
plt.subplot(1, 2, 1) 
plt.hist(data_set_1, bins=15, color='teal', edgecolor='black')
plt.title("A) Uniform Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

# --- Second Subplot (1, 2, 2) ---
plt.subplot(1, 2, 2) 
plt.bar(categories_2, data_set_2, color='magenta')
plt.title("B) Discrete Count Bar Chart")
plt.xlabel("Category")
plt.ylabel("Count")

plt.tight_layout() # Adjusts subplot parameters for a tight layout
plt.show()
```

```python
### Assignment 2: Histogram with lines

'''
Using the function numpy.random.exponential(). Use it to create a NumPy array of length 500 named 'data_exp'
with a scale parameter of two. Plot a histogram of this data, adding vertical lines to clearly indicate the
median and the 90th percentile of the data_exp array. Ensure the plot has a clear title, axis labels,
and a legend identifying the line markers. Again brownie points if you manage to customize this plot to look good.
'''
def assignment2():
    data_exp = np.random.exponential(scale=2, size=500)
    
    plt.figure(figsize=(10, 6))
    plt.hist(data_exp, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)

    median = np.median(data_exp)
    percentile_90 = np.percentile(data_exp, 90)

    plt.axvline(median, color='blue', linestyle='--', label=f'Median: {median:.2f}')
    plt.axvline(percentile_90, color='red', linestyle='-.', label=f'90th Percentile: {percentile_90:.2f}')

    plt.title("Histogram of Exponential Data with Median and 90th Percentile")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()
assignment2()  
```
